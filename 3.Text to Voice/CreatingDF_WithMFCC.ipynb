```python
import librosa
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display as ipd
import librosa.display
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_csv(r'D:\Capstone\Creating datasets\Data_frame_Output.csv')
```

```python
def feature_extractor(file,list):
    audio , sample_rate1 = librosa.load(file)
    mfcc_fetures = librosa.feature.mfcc(y=audio,sr=sample_rate1,n_mfcc=40)
    mfcc_scaled_features = np.mean(mfcc_fetures.T,axis=0)
    list.append(mfcc_scaled_features)
    
```

```python
def feature_extractor2(file, feature_list):
    audio, sample_rate = librosa.load(file)
    mfcc_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    
    # Calculate mean and standard deviation along the time axis
    mean_mfcc = np.mean(mfcc_features, axis=1, keepdims=True)
    std_mfcc = np.std(mfcc_features, axis=1, keepdims=True)
    
    # Z-score normalization
    normalized_mfcc = (mfcc_features - mean_mfcc) / std_mfcc
    
    # Take the mean along the time axis and append to the list
    mean_normalized_mfcc = np.mean(normalized_mfcc.T, axis=0)
    feature_list.append(mean_normalized_mfcc)
```

```python
mfcc_feature_list = []
for path in dataframe['Path']:
    feature_extractor(path,mfcc_feature_list)
mfcc_feature_list    
```

```python
mfcc_feature_list = []
for path in dataframe['Path']:
    feature_extractor2(path,mfcc_feature_list)
mfcc_feature_list    
```

```python
dataframe['Mfcc_Fratures'] = mfcc_feature_list
```

```python
dataframe.to_csv(r'DataFrame_with_Mfcc.csv', index=False)
```
