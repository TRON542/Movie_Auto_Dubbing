```python
import os
import pandas as pd
import assemblyai as aai
import time
aai.settings.api_key = "fb7e62fc83f64fa2b45e7b97c13ee6a0"
```

```python
def audio_to_text(File_path,audio_text):
    FILE_URL = File_path
    print(FILE_URL)
    transcriber = aai.Transcriber()
    transcript = transcriber.transcribe(FILE_URL)
    final_txt = transcript.text
    audio_text.append(final_txt)
    
```

```python
def taking_paths(audio_loc,audio_path_list):
    size = len(os.listdir(audio_loc))
    for segment_number in range(1, size + 1):
        segment_path = f'segment_{segment_number}.wav'
        path = os.path.join(audio_loc, segment_path)
        audio_path_list.append(path)
```

```python
audio_path_list = []
main_folder = r'D:\Capstone\VoiceDatasets\Output\Noise removed'
for sub_folder in os.listdir(main_folder):
    segment_path = f'{main_folder}\\{sub_folder}'
    temp = []
    taking_paths(segment_path,temp)
    temp2 = temp.copy()
    audio_path_list.append(temp2)
    temp.clear()
```

```python
audio_text = []
for i in audio_path_list:
    temp = [] 
    for j in i:
        audio_to_text(j,temp)
    temp2 = temp.copy()
    audio_text.append(temp2)    
```
